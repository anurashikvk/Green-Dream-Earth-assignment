```python
import os
import csv
import requests
from bs4 import BeautifulSoup
from lxml import html
import re

url="https://www.lse.ac.uk/study-at-lse/Undergraduate/degree-programmes-2024/BA-History?_gl=1*1j8om2u*_ga*MjA5MDU5MzE3My4xNzEwODQ0Nzky*_ga_LWTEVFESYX*MTcxMDg0NDc5Mi4xLjEuMTcxMDg0NDgzMC4yMi4wLjA."
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    tree = html.fromstring(response.content)

    # Extracting degree type
    li_element = soup.find('li', class_='keyDetails__item--grad')
    degree_type = li_element.text.strip()

    # Extracting application deadline
    application_deadline = soup.find_all('td')[1].text

    # Extracting duration
    duration = soup.find_all('td')[2].text

    # Extracting tuition fee
    elements_fee = tree.xpath('//*[@id="form1"]/div[3]/div/div[2]/section/div/section[4]/div/div/p[5]')
    for element in elements_fee:
        text_content = element.text.strip()
        pattern = r'£(\d{1,3}(?:,\d{3})*(?:\.\d{1,2})?)'
        matches = re.search(pattern, text_content)
        tuition_fee = "£"+matches.group(1)

    # Extracting description
    description = tree.xpath('//*[@id="form1"]/div[3]/div/div[2]/div/p[1]')
    for element in description:
        description_brief = element.text.strip()

    # Extracting requirements
    requirements_elements = tree.xpath('//*[@id="form1"]/div[3]/div/div[2]/section/div/section[2]/div/div/p[position() >= 2 and position() <= 6] | //*[@id="form1"]/div[3]/div/div[2]/section/div/section[2]/div/div/ul')
    requirements = '\n'.join([element.text_content().strip() for element in requirements_elements])

    # Storing data in CSV
    with open('degree_details.csv', mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Degree Type', 'Application Deadline', 'Duration', 'Tuition Fee', 'Description', 'Requirements'])
        writer.writerow([degree_type, application_deadline, duration, tuition_fee, description_brief, requirements])
else:
    print(f"Failed to fetch the page: {url}")
```

```python
import requests
import re
from textblob import TextBlob
from collections import Counter

# Fetch webpage content
url = "https://www.lse.ac.uk/study-at-lse/Undergraduate/degree-programmes-2024/BA-History?_gl=1*1j8om2u*_ga*MjA5MDU5MzE3My4xNzEwODQ0Nzky*_ga_LWTEVFESYX*MTcxMDg0NDc5Mi4xLjEuMTcxMDg0NDgzMC4yMi4wLjA."
response = requests.get(url)
html_content = response.text

# Extract text using regular expressions
text = re.findall(r'<p>(.*?)</p>', html_content, re.DOTALL)

# Concatenate paragraphs into a single string
text = "\n".join(text)

# Print extracted text
print(text)

# Sentiment analysis
blob = TextBlob(text)
sentiment = blob.sentiment
print("\nSentiment Analysis:")
print("Polarity:", sentiment.polarity)
print("Subjectivity:", sentiment.subjectivity)

# Tokenization and word frequency
words = re.findall(r'\b\w+\b', text.lower())
word_freq = Counter(words)

# Print most common words
print("\nMost common words:")
print(word_freq.most_common(10))
```

```python
import requests
from bs4 import BeautifulSoup
import spacy

# Load English language model
nlp = spacy.load("en_core_web_sm")

# Fetch webpage content
url = "https://www.lse.ac.uk/study-at-lse/Undergraduate/degree-programmes-2024/BA-History?_gl=1*1j8om2u*_ga*MjA5MDU5MzE3My4xNzEwODQ0Nzky*_ga_LWTEVFESYX*MTcxMDg0NDc5Mi4xLjEuMTcxMDg0NDgzMC4yMi4wLjA."
response = requests.get(url)
html_content = response.text

# Parse HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Extract text
text = soup.get_text()

# Process text with spaCy
doc = nlp(text)

# Initialize variables to store extracted information
degree_type = None
application_deadline = None
duration = None
entry_requirements = None
overseas_fees = None
degree_description = None

# Iterate through sentences to extract information
for sent in doc.sents:
    sent_text = sent.text.strip()
    if "undergraduate" in sent_text.lower() or "postgraduate" in sent_text.lower():
        degree_type = sent_text
    elif "application deadline" in sent_text.lower():
        application_deadline = sent_text
    elif "duration" in sent_text.lower():
        duration = sent_text
    elif "entry requirements" in sent_text.lower():
        entry_requirements = sent_text
    elif "overseas student fees" in sent_text.lower():
        overseas_fees = sent_text
    elif "brief degree description" in sent_text.lower():
        degree_description = sent_text

# Print extracted information
print("Degree Type:", degree_type)
print("Application Deadline:", application_deadline)
print("Duration:", duration)
print("Admission Eligibility/Entry Requirements:", entry_requirements)
print("Overseas Student Fees:", overseas_fees)
print("Brief Degree Description:", degree_description)
```
